```python
#1: lam quen voi class => ham khoi tao 
class student:
    def __init__(self, name, age):
        self.name = name
        self.age = age

stu = student("Trinh Minh Khanh", 18)
print(stu.name)
print(stu.age)
```

```python
#2 : lam quen voi class and attribute
#de bai: In ra man hinh dien tich, chu vi hinh chu nhat
class rectangle:
    def __init__(self, height, width):
        self.height = height
        self.width = width

obj1 = rectangle(20, 5)
print("Dien tich cua hinh chu nhat la: ", obj1.height * obj1.width)
print("Chu vi cua hinh chu nhat la: ", (obj1.height + obj1.width) * 2)

        
```

```python
#3 : lam quen voi class and attribute
#de bai: in ra ten , neu khong truyen vao tham so ten thi in ra Unknown
class person:
    def __init__(self, name = "Unknown"):
        self.name = name

person1 = person()
print(person1.name)
        
```

```python
#4: lam quen voi class and method
class dog:
    def __init__(self, name):                      #Ham khoi tao voi thuoc tinh la name
        self.name = name
    
    def bark(self):                                #Ham nay tro vao 1 doi tuong
        print("Gau gau , toi la : ", self.name)


dog1 = dog("KOKO")
dog.bark(dog1)
```

```python
#5: May tinh cam tay
class caculator:
    def __init__(self, number1, number2):
        self.number1 = number1
        self.number2 = number2
    
    def add(self):
        print("Ket qua cua phep cong la : ", self.number1 + self.number2)
    def sub(self):
        print("Ket qua cua phep tru la : ", self.number1 - self.number2)
    def mul(self):
        print("Ket qua cua phep nhan la : ", self.number1 * self.number2)
    def div(self):
        if self.number2 != 0:
            print("Ket qua cua phep chia la : ", self.number1 / self.number2)
        else:
            print("Khong chia duoc")
    
cal = caculator(5, 0)
caculator.add(cal)
caculator.sub(cal)
caculator.mul(cal)
caculator.div(cal)
```

```python
#6: Tinh diem trung binh va xep hang hoc sinh
class student:
    def __init__(self, name, math, phys, chem):
        self.name = name
        self.math = math
        self.phys = phys
        self.chem = chem
    
    def avg(self):
        return (self.math + self.phys + self.chem) / 3
    def rank(self):
        if student.avg(self) >= 8.0:
            print("Hoc sinh gioi")
        elif 6.5 <= student.avg(self) < 8.0 :
            print('Hoc sinh kha')
        elif 5.0 <= student.avg(self) < 6.5:
            print("Hoc sinh trung binh")
        else:
            print("Hoc sinh yeu")

stu = student("Trinh Minh Khanh", 5, 5, 5)
student.rank(stu)
    
```
