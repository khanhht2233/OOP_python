```python
#1: Tao lop ke thua  => 
class Information:
    learn = "python"
    def __init__(self, para_name, para_age, para_gender, para_university):
        self.name = para_name
        self.age = para_age
        self.gender = para_gender
        self.university = para_university

    def sublects(self):
        return ["DSTT", "GT1", "TDTT", "TTHCM"]
    

stu1 = Information("Trinh Minh Khanh", "18", "Male", "Univesity of Engineering and Technology")

class Inheritance(Information):
    def __init__(self, para_name, para_age, para_gender, para_university, para_adress):  #khoi tao ham moi co them thuoc tinh para_adress 

        super().__init__(para_name, para_age, para_gender, para_university)              #ke thua contsuctor : tai su dung ham cu bang ham super().__init__()

        self.adress = para_adress                                                        #cap nhat tham so moi them

in_stu1 = Inheritance('Sinh vien', '19', "Female", 'VietNam National University', "Nghe An")
print(in_stu1.name, in_stu1.age, in_stu1.gender, in_stu1.university, in_stu1.adress, sep = "\n")
print()
print("\n".join(in_stu1.sublects()))
```

```python
#Van dung class method
#__init__ : khoi tao thuoc tinh objects
#classmethod khong co self chi co cls : cls chi cho phep truy cap duoc attribute( du lieu chung cua class ) cua class khong cho phep truy cap vao thuoc tinh cua objects
# => khong the can thiep duoc vao du lieu cua objects


class Teacher:
    school = "UET"
    def __init__(self, name, age):
        self.name = name
        self.age = age
    
    @classmethod
    def change_school(cls, school):
        cls.school = school

t1 = Teacher("LE NGUYEN KHOI", 45,)
print(t1.name, t1.age, Teacher.school, sep = "\n") #  su dung attribute nguyen ban cua class 

t1.change_school("HUSt")                           # can thiep thay doi du lieu 
print(t1.name, t1.age, t1.school, sep = "\n") 
        
```

```python
#2 : basic
# special methods : la nhung ham duoc su dung trong OOP co vai tro giong nhu nhung ham built_in trong lap trinh binh thuong
class Book:
    def __init__(self, title, pages):
        self.title = title
        self.pages = pages

    def __str__(self):
        return f"Book : {self.title} ({self.pages} pages)"
    
    def __len__(self):
        return self.pages

b1 = Book("Nhung ngay tho au", 1000)    
print(b1)
print(len(b1))
    
```

```python
#3 : So sanh objects
class Student:
    def __init__(self, name, score):
        self.name = name
        self.score = score

    def __str__(self):
        return f'{self.name}, {self.score}'
    
    def __eq__(self, other):
        return self.score == other.score

    def __gt__(self, other):
        return self.score > other.score

s1 = Student("Khanh", 8.5)
s2 = Student("An", 7.0)

print(s1)
print(s1 == s2)   # False
print(s1 > s2)    # True
 
```

```python
#4 : toan tu '+'
class Vector:
    def __init__(self, x):
        self.x = x

    def __add__(self, other):
        return Vector(self.x + other.x)

    def __str__(self):
        return f"{self.x}"

v1 = Vector(1)
v2 = Vector(2)
v3 = v1 + v2
print(v3)
        
        
```
