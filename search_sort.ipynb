```python
#1 : linear search
def ls(arr, n):
    for i in range(len(arr)):
        if arr[i] == n:
            return n 
    return -1
```

```python
#2 : Binary search => tim kiem tren mang da sap xep
def bs(arr, n):
    left, right = 0, len(arr) - 1
    while left <= right:
        mid = (left +right) // 2
        if arr[mid] == n:
            return mid
        elif arr[mid] > n :
            right = mid - 1
        else:
            left = mid + 1
    return -1
```

```python
#3 : Bubble sort (sap xep noi bot)
def bubble_sort(arr, n):
    for i in range(n - 1):
        status = False
        for j in range(n - i - 1):
            if arr[j] > arr[j + 1]:
                arr[j], arr[j + 1] = arr[j + 1], arr[j]
                status = True
        if not status:
            break
    return arr
```

```python
#4 : selection sort (sap xep chon)
def selection_sort(arr, n):
    for i in range(n - 1):
        min_idx = i
        for j in range(i + 1, n):
            if arr[j] < arr[min_idx]:
                min_idx = j
        arr[min_idx], arr[i] = arr[i], arr[min_idx]
    return arr
```

```python
#5 : insertion sort (sap xep chen)
def insertion_sort(arr, n):
    for i in range(1, n):
        j, key = i - 1, arr[i]
        while j >= 0 and arr[j] > key:
            arr[j + 1] = arr[j]
            j -= 1
        arr[j + 1] = key
    return arr
```

```python
#6 merge sort (sap xep tron)
def merge(arr):
    if len(arr) <= 1:
        return arr

    mid = len(arr) // 2
    left = merge(arr[:mid])
    right = merge(arr[mid :])

    return merge_sort(left, right)

def merge_sort(left, right):
    i, j = 0, 0
    res = []
    while i < len(left) and j < len(right):
        if left[i] < right[j] :
            res.append(left[i])
            i += 1
        else:
            res.append(right[j])
            j += 1
    res.extend(left[i:])
    res.extend(right[j:])
    return res


ls = [9, 6, 7, 8, 1 ,3 ,2]
n = len(ls)
print(merge(ls))
```

```python
#7 : quick sort
def partition(arr, low, high):
    pivot = arr[high]
    i = low - 1
    for j in range(low, high):
        if arr[j] <= pivot:
            i += 1
            arr[i], arr[j] = arr[j], arr[i]
    arr[i + 1], arr[high] = arr[high], arr[i + 1]
    return i + 1

def quick_sort(arr, low, high):
    if low < high:
        mid = partition(arr, low, high) 
        quick_sort(arr, low, mid - 1)
        quick_sort(arr, mid + 1, high)

arr = [9, 6, 7, 8, 1 ,3 ,2]
low, high = 0, len(arr) - 1
quick_sort(arr, low, high)
print(arr)
```
