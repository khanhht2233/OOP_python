```python
def tinh_chiet_khau(gia, phan_tram):
    return gia * (1 - phan_tram)

# Ứng dụng assert để test các kịch bản
assert tinh_chiet_khau(100, 0.1) == 90.0
assert tinh_chiet_khau(200, 0.1) == 100.0
# Nếu hàm chạy sai, chương trình sẽ dừng tại đây và báo lỗi
```

```python
def giai_thua(n):
    res = 1
    while n != 0:
        res *= n
        n -= 1
    return res

assert giai_thua(1) == 1, "check1"
assert giai_thua(0) == 0, 'check2'
```
